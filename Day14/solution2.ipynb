```python
#Import the grid
with open("input.txt") as f:
    grid = tuple(line.strip() for line in f)
print(f'imported grid: {grid}')
```

```python
# define a cycle funtion

def cycle():
    global grid
    for _ in range(4):
        grid = tuple(map("".join, zip(*grid)))
        grid = tuple("#".join(["".join(sorted(list(group), reverse = True)) for group in row.split("#")])for row in grid)
        grid = tuple(row[::-1] for row in grid)
    

# look for an already seen configuration to avoid calling 1B times the function "cycle()"

seen = {grid}
array = [grid]
iter = 0

while True:
    iter += 1
    cycle()
    if grid in seen:
        break
    seen.add(grid)
    array.append(grid)
        
first = array.index(grid)

grid = array[(1000000000 - first) % (iter - first) + first]
        
print(sum(row.count("O") * (len(grid) - r) for r, row in enumerate(grid)))

'''from https://github.com/hyper-neutrino/advent-of-code/blob/main/2023/day14p2.py'''
```
